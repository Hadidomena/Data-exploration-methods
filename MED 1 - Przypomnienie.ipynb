* państwa o zbliżonej liczbie ludności do Polski
* najbardziej podobne do Polski państwa świata pod względem zarówno powierzchni jak i liczby ludności

```python
polska = panstwa[panstwa['Country']=='Poland']

# Liczba ludności Polski
ludnoscPl = polska.iloc[0,2]
powPL = polska.iloc[0,3]
print(f"Liczba ludności Polski: {ludnoscPl}")
print(f"Powierzchnia Polski: {powPl}\n")

print("Podejście do szukania podobnych państw w narzuconych zakresach wartości\n")
# Państwa o zbliżonej liczbie ludności do Polski
czescLudnosci = 12
delta_ludnosc = ludnoscPl / czescLudnosci  # tolerancja +/- 1/10 populacji Polski
zblizona_ludnosc = panstwa[(panstwa.iloc[:,2] > (ludnoscPl - delta_ludnosc)) & 
                          (panstwa.iloc[:,2] < (ludnoscPl + delta_ludnosc))]

print(f"Państwa o zbliżonej liczbie ludności do Polski (+/- 1/{czescLudnosci} populacji Polski):")
print(zblizona_ludnosc[['Country', 'Population 2025']])

# Najbardziej podobne państwa pod względem powierzchni i liczby ludności
czesc_powierzchni = 8
delta_powierzchnia = powPL / czesc_powierzchni  # tolerancja +/- 1/10 powierzchni
delta_ludnosc_podobne = delta_ludnosc # tolerancja stała

podobne_panstwa = panstwa[(panstwa.iloc[:,3] > (powPl - delta_powierzchnia)) & 
                         (panstwa.iloc[:,3] < (powPl + delta_powierzchnia)) &
                         (panstwa.iloc[:,2] > (ludnoscPl - delta_ludnosc_podobne)) & 
                         (panstwa.iloc[:,2] < (ludnoscPl + delta_ludnosc_podobne))]

print(f"Najbardziej podobne państwa do Polski (powierzchnia +/- 1/{czesc_powierzchni} powierzchni Polski, ludność +/- 1/{czescLudnosci} populacji Polski):")
print(podobne_panstwa[['Country', 'Population 2025', 'Land Area (KmÂ²)']])

print("\nAlternatywne podejście oparte na obliczeniu podobieństwa ogólnego i wymienieniu najbliższych Polsce państw\n")
panstwa_copy = panstwa.copy()
panstwa_copy['podobienstwo_ludnosc'] = abs(panstwa_copy['Population 2025'] - ludnoscPl) / ludnoscPl
panstwa_copy['podobienstwo_powierzchnia'] = abs(panstwa_copy['Land Area (KmÂ²)'] - powPl) / powPl
panstwa_copy['podobienstwo_ogolne'] = (panstwa_copy['podobienstwo_ludnosc'] + 
                                      panstwa_copy['podobienstwo_powierzchnia']) / 2

ilosc_wymienionych = 4
najbardziej_podobne = panstwa_copy.nsmallest(ilosc_wymienionych + 1, 'podobienstwo_ogolne')[['Country', 'Population 2025', 
                                                                         'Land Area (KmÂ²)', 'podobienstwo_ogolne']]
print(f"{ilosc_wymienionych} najbardziej podobnych państw do Polski (wykluczając Polskę):")
print(najbardziej_podobne[najbardziej_podobne['Country'] != 'Poland'])
```

### Wnioski odnośnie typów wykresów

*   **Histogram**
    *   Jest wygodny, jeśli chodzi o poszukiwanie przedziałów, w których znajduje się większość danych.
    *   Ułatwia generalne badanie rozkładu danych dla pojedynczego atrybutu.

*   **Wykres Pudełkowy (Box Plot)**
    *   Jest wygodniejszy przy poszukiwaniu mediany i typowych wartości jednego atrybutu w jego całkowitym zakresie.
    *   Po dodaniu podziału na klasy ułatwia obserwację rozkładu atrybutu względem tych klas.
    *   Pozwala na szybkie zaobserwowanie wartości maksymalnych, minimalnych i najczęstszych (kwartyle) wewnątrz danej klasy.

*   **Wykres Punktowy (Scatter Plot)**
    *   Pozwala na przybliżone zauważenie zależności jednego atrybutu względem drugiego.
    *   Jest przydatny przy badaniu tych zależności.
    *   Przy użyciu podziału na klasy pozwala na wygodne badanie zależności klas od dwóch atrybutów jednocześnie.

```python
import math

print("Porównanie mediany i średniej atrybutów dla zbioru iris")
numeric_columns = iris.select_dtypes(include=['number'])
for c in numeric_columns:
    diff = iris[c].mean()-iris[c].median()
    abs_diff = math.fabs(diff)
    print(f"Różnica między średnią a medianą dla atrybutu {c} wynosi {diff}" +
      f"\na więc {abs_diff / iris[c].median() * 100}% mediany" + 
      f"\noraz {abs_diff / iris[c].mean() * 100}% średniej")
    
numeric_columns = panstwa.select_dtypes(include=['number']).iloc[:,1:]
print("Porównanie mediany i średniej dla zbioru państwa")
for c in numeric_columns:
    diff = panstwa[c].mean()-panstwa[c].median()
    abs_diff = math.fabs(diff)
    print(f"Różnica między średnią a medianą dla atrybutu {c} wynosi {diff}" +
      f"\na więc {abs_diff / panstwa[c].median() * 100}% mediany" + 
      f"\noraz {abs_diff / panstwa[c].mean() * 100}% średniej")
    
```

### Różnice między medianą a średnią jako wskaźnik rozkładu danych

Porównanie mediany i średniej pozwala wywnioskować, czy rozkład wyników jest równomierny.

*   **Gdy średnia i mediana są do siebie zbliżone**, sugeruje to **symetryczny rozkład danych** względem średniej.
*   **Gdy średnia i mediana różnią się od siebie**, sugeruje to **niesymetryczność wyników** wobec średniej. Może to oznaczać obecność wyników znacznie większych/mniejszych od średniej lub większą częstość wyników o wartości wyższej/niższej od średniej.

> **Zadanie 5** Skonfrontuj macierz korelacji z wykresami punktowymi zbiorów `d` oraz `iris`. Jaka jest zależność między konfiguracją punktów na wykresie punktowym a wartością miary korealcji ? Jakie wnioski odnoszące się do poszczególnych atrybutów można wyciągnąć na podstawie analizy macierzy korelacji ?  

```python
print("Macierz korelacji dla zbioru d:")
print(d.select_dtypes(include=[np.number]).corr())

print("Macierz korelacji dla zbioru iris:")
print(iris.select_dtypes(include=[np.number]).corr())

plt.figure(figsize=(12, 8))

#Wykresy punktowe dla d ułatwiające analizę
plt.subplot(2, 2, 1)
plt.scatter(d['atrybut1'], d['atrybut2'])
plt.xlabel('atrybut1')
plt.ylabel('atrybut2')
plt.title('Zbiór d - wszystkie klasy')

plt.subplot(2, 2, 2)
for klasa in d['klasa'].unique():
    subset = d[d['klasa'] == klasa]
    plt.scatter(subset['atrybut1'], subset['atrybut2'], label=klasa)
plt.xlabel('atrybut1')
plt.ylabel('atrybut2')
plt.title('Zbiór d - podział na klasy')
plt.legend()

# Zastosowanie heatmap jako ułatwionej wizualizacji korelacji
# Heatmapa korelacji dla zbioru d
plt.subplot(2, 2, 3)
sns.heatmap(d.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Heatmapa korelacji - zbiór d')

# Heatmapa korelacji dla zbioru iris
plt.subplot(2, 2, 4)
sns.heatmap(iris.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Heatmapa korelacji - zbiór iris')

plt.tight_layout()
plt.show()
```

### Wnioski z analizy korelacji

#### 1. Ogólne zasady interpretacji korelacji

*   **Siła korelacji:** Im wyższa wartość bezwzględna korelacji (bliższa 1 lub -1), tym punkty na wykresie rozrzutu układają się bardziej wzdłuż linii prostej, co wskazuje na silniejszą **zależność liniową**.
*   **Brak korelacji liniowej:** Korelacja bliska 0 oznacza brak liniowej zależności. Punkty są rozrzucone chaotycznie, co sugeruje, że atrybuty nie są ze sobą powiązane w sposób liniowy.
*   **Korelacja dodatnia:** Punkty układają się wzdłuż rosnącej prostej. Wzrostowi wartości jednego atrybutu towarzyszy wzrost wartości drugiego.
*   **Korelacja ujemna:** Punkty układają się wzdłuż malejącej prostej. Wzrostowi wartości jednego atrybutu towarzyszy spadek wartości drugiego.
*   **Korelacja a przyczynowość:** Znaczna wartość korelacji wskazuje na zależność statystyczną, ale **niekoniecznie na związek przyczynowo-skutkowy**. Atrybuty mogą:
    *   Być od siebie zależne (np. w zbiorze prostokątów o stałej powierzchni, szerokość i długość będą skorelowane ujemnie).
    *   Mieć wspólną, ukrytą przyczynę.
    *   Być skorelowane czysto przypadkowo.

*   **Brak korelacji:** Bardzo niska wartość korelacji między szerokością a długością kielicha sugeruje brak liniowego związku między tymi cechami.
*   **Wysoka dodatnia korelacja:**

    *   **Sugestia:** Kwiaty o dłuższych kielichach mają tendencję do posiadania zarówno dłuższych, jak i szerszych płatków.
